```python
num3 = 0
num4 = 0
for idx in range(50):
  img_path = os.listdir("./Dataset/test_images_2")[idx]
  input_path = read_image(f'{"./Dataset"}/test_images_2/{img_path}') / 255
  output_path = read_image(f'{"./Dataset"}/test img 2/{img_path}') / 255

  print("input:",input_path.shape)
  print("output:",output_path.shape)
  print()

  if input_path.shape[0] == 3:
    num3 += 1
  elif input_path.shape[0] == 4:
    num4 += 1

print(num3)
print(num4)
```
